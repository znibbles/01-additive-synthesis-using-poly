```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
imp = signal.unit_impulse(256, 128)
n = np.arange(-128, 128)


train = np.tile(imp, 8)
plt.plot(train)
plt.show()

plt.plot(np.cumsum(train))
plt.show()
```

```python
n = np.arange(-128, 128)
bli = np.sinc(n/2)
plt.plot(bli)
plt.show()

blit = np.tile(bli, 8)
plt.plot(blit)
plt.show()

plt.plot(np.cumsum(blit)/2)
plt.show()
```
