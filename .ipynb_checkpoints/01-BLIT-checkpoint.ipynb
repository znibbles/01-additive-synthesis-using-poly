```python
def tfPlot(sig, fs=44100, timeLen=None):
    """
    Function for getting a time/frequency domain plot.
    """
    N = len(sig)
    t = np.arange(N)/fs
    plt.figure(figsize=[17,5])
    plt.subplot(121)
    plt.title('Time Domain')
    plt.plot(t,sig)
    if timeLen !=None:
        plt.xlim([0,timeLen])
    plt.grid()
    plt.xlabel('Time[sec]')
    plt.subplot(122)

    plt.semilogx(np.linspace(1,fs/2, int(N/2)), abs(np.fft.fft(sig))[0:int(N/2)])
    plt.xlabel('Freq[Hz]')
    plt.title('Spectrum')
    plt.grid()
    plt.show()

```

# Naive Sawtooth

- will inevitably lead to aliasing at higher frequencies

```python
# naive sawtooth

def makeNaiveSaw(f1=100):
    global sig
    T = 0.3 #sec
    fs = 44100. #Hz
    N = T*fs #num Samples
    n = np.arange(N)+1. #sample Index
    t = n/fs #time

    freq = f1 #Hz
    P = 1/freq * fs # (1/freq)*fs
    
    y = np.zeros(int(N))

    for i in range(len(n)):
        y[i] = 2* ((i % P)/P - 0.5)

    tfPlot(y, timeLen=4*P/fs)
    sig = y
```

# Bandlimited Impulse Trains (BLITs)
A BLIT signal, $y(n)$ with $P$ as Period in Samples is given by:

$$ y(n) = (M/P) Sinc_M[(M/P)n] $$
where


$$ Sinc_M(x) = \frac{sin( \pi x)}{M sin(\pi x /M)}$$

$M$ = Number of Harmonics

Calculated via
$$ M = 2\cdot floor(P/2)+1 $$

## Issues:
- the denominator of `sincM` will be 0 if $n$ is an integer factor of $P$
- the numerator of `sincM` will be 0 if $x$ is an integer, which will be the case when $M = P$
- Both Filters for the Realtime Sawtooth probably need to be slightly frequency dependent.


```python
def makeSimpleImpulseTrain(f1=100):
    global sig
    T = 0.3 #sec
    fs = 44100. #Hz
    N = T*fs #num Samples
    n = np.arange(N)+1. #sample Index
    t = n/fs #time

    freq = f1 #Hz
    P = 1/freq * fs # (1/freq)*fs
    
    y = np.zeros(int(N))

    for i in range(len(n)):
        y[i] = 1 if np.floor(i % P) == 0 else 0

    tfPlot(y, timeLen=4*P/fs)
    sig = y
    
```

```python
# v = interactive(makeSimpleImpulseTrain, f1=(20.0,5000.0),)
# display(v)

makeSimpleImpulseTrain(2000)
```

```python
Audio(data=sig, rate=44100)
```

```python
avg = np.average(sig) #get DC component 
simple_saw = np.cumsum(sig-avg) # integrate after removing the dc component
simple_saw -= np.average(simple_saw) #remove DC component of result

tfPlot(simple_saw, fs=44100, timeLen=0.005)
```

```python
Audio(data=simple_saw, rate=44100)
```

```python
def sincM(x, M):
    return np.sin(np.pi*x)/(M*np.sin((np.pi*x)/M))
def sampledSinc(n,M,P):
    output = np.zeros_like(n)
    
    if M==P:
        cond = np.zeros_like(n, dtype=bool)
        condInv = np.ones_like(n, dtype=bool)
    else:    
        cond = ((n / P) % 1 != 0)
        condInv = ((n / P) % 1 == 0)

    output[cond] = (M/P)*sincM((M/P)*n[cond], M)
    P-=0.001
    output[condInv] = (M/P)*sincM((M/P)*n[condInv], M)
    return output
```

```python
def makeBLIT(f1=100):
    global sig
    T = 0.1 #sec
    fs = 44100. #Hz
    N = T*fs #num Samples
    n = np.arange(N)+1. #sample Index
    t = n/fs #time

    freq = f1 #Hz
    T1 = 1./freq #period in Seconds
    Ts = 1./fs #sampling Period
    P = T1/Ts # (1/freq)*fs
    
    M = 2*np.floor(P/2.)+1 #Number of Harmonics
    y = sampledSinc(n, M, P)
    tfPlot(y, timeLen=4*P/fs)
    sig = y
```

```python
makeBLIT(2000)
```

```python
Audio(data=sig, rate=44100)
```

# Realtime Capable Sawtooth

For the realtime Sawtooth the DC component is approximated via an IIR filter and the integration is done via a Leaky integrator (one pole). 


## Leaky Integrator 

$$ H(z) = \frac{1}{1 -z^{-1}\alpha}$$

where $\alpha$ is very close to 1.

This basically means
$$ y(n) = x(n)+y(n-1)\alpha $$



```python
def leakyIntegrator(x, alpha=1-10e-6):
    return signal.lfilter([1.0, 0.], [1., -alpha], x)    
```

```python
def leakyIntegratorPython(x, alpha=1-10e-6):
    """
    This just demaonstrates what happens inside the much more efficient scipy version.
    """
    Y = np.zeros_like(x)
    y1 = 0
    for i in range(len(x)):
        y = x[i]*1+y1*alpha
        y1=y
        Y[i]=y
    return Y
```

```python
N = 30
alpha = 0.99
testIn = signal.unit_impulse(N, int(N/2))
testOut = leakyIntegrator(testIn, alpha=alpha)
testOutP = leakyIntegratorPython(testIn, alpha=alpha)

plt.stem(np.arange(N),testOut)
plt.title('scipy')
plt.show()
plt.stem(np.arange(N),testOutP)
plt.title('python')
plt.show()
```

## Plot of Different $\alpha$ Coefficiants

```python
colors = ['ro', 'go', 'bo']
for i in range(3):
    alpha = 1-10**-(i*0.5+1)
    Y = leakyIntegrator(testIn, alpha=alpha)
    plt.stem(np.arange(N),Y, 'g',label= 'alpha = %f' % alpha, markerfmt=colors[i])
plt.legend(loc='best')
plt.show()
```

```python
def runningAvg(x, alpha=0.995):
    """
    IIR aproximation of a running average
    """
    return signal.lfilter([1-alpha,0],[1,-alpha],x)
```

Having all that, we still use the BLIT from above as $y$:

```python
newAvg = runningAvg(sig) #get DC component of BLIT
print(np.average(newAvg))
saw = leakyIntegrator(sig-newAvg) # integrate after removing the dc component
saw -= runningAvg(saw) #remove DC component of result

tfPlot(saw, fs=fs)
```

## Sweep

```python
T = 5 #sec
fs = 44100. #Hz
N = int(T*fs) #num Samples
n = np.arange(N)+1 #sample Index
t = n/fs #time
freqSweep = np.linspace(20, 10000, N)

Y = np.zeros_like(t)


for i in range(N):
    freq = freqSweep[i] #Hz
    T1 = 1./freq #period in Seconds
    Ts = 1./fs #sampling Period
    P = T1/Ts # (1/freq)*fs
    M = 2*np.floor(P/2.)+1 #Number of Harmonics
    y = sampledSinc(np.array([i]), M, P)
    Y[i]=y

plt.figure(figsize=[17,5])
plt.specgram(Y, Fs=fs)
plt.show()
    
```

```python
plt.plot(Y)
plt.show()
```

```python
Audio(data=Y, rate=fs)
```
