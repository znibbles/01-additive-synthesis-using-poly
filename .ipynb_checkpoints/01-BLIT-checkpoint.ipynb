$$ y(n) = (M/P) Sinc_M[(M/P)n] $$
where


$$ Sinc_M(x) = \frac{sin( \pi x)}{M sin(\pi x /M)}$$

$M$ = Number of Harmonics

Calculated via
$$ M = 2[P/2]+1 $$

$P$ = Period in Samples

```python
def SincM(x, M):
    return np.sin(np.pi*x)/(M*np.sin((np.pi*x)/M))
def sampledSinc(n,M,P):
    return (M/P)*SincM((M/P)*n, M)
```

```python
T = 10 #sec
fs = 8000 #Hz
N = T*fs #num Samples
n = np.arange(N)+1 #sample Index
t = n/fs #time
```

```python
freq = 100 #Hz
T1 = 1/freq #period in Seconds
Ts = 1/fs #sampling Period
M = 2*(P/2)+1 #Number of Harmonics
P = T1/Ts # (1/freq)*fs
y = sampledSinc(n, M, P)
```

```python
plt.figure(figsize=[17,5])
plt.subplot(121)
plt.plot(t,y)
plt.xlabel('Time[sec]')
plt.subplot(122)

plt.semilogx(np.linspace(1,fs/2, int(N/2)), abs(np.fft.fft(y))[0:int(N/2)])
plt.xlabel('Freq[Hz]')
plt.title('Spectrum')
plt.show()
```
